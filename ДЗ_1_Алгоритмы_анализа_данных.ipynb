```python
import numpy as np
```

**1. –ü–æ–¥–±–µ—Ä–∏—Ç–µ —Å–∫–æ—Ä–æ—Å—Ç—å –æ–±—É—á–µ–Ω–∏—è (eta) –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π**

```python
# –ó–∞–¥–∞—á–∞: –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –±–∞–ª–ª–æ–≤ –ï–ì–≠ —É—á–µ–Ω–∏–∫–∞ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –∫–æ–ª-–≤–∞ –ª–µ—Ç —Å—Ç–∞–∂–∞ –µ–≥–æ —Ä–µ–ø–µ—Ç–∏—Ç–æ—Ä–∞
```

```python
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
```

```python
X.shape
```

```python
y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
```

```python
# –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–π –æ—à–∏–±–∫–∏:
```

```python
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

```python
# –ò–º–µ–µ—Ç—Å—è –º–æ–¥–µ–ª—å, –æ–ø—Ä–µ–¥–µ–ª—è—é—â–∞—è "–≤–µ—Å–∞" w –¥–ª—è –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–π x –º–µ—Ç–æ–¥–æ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞:
```

```python
n = X.shape[0]

# eta = 1e-2
# n_iter = 100 - –ü–ê–†–ê–ú–ï–¢–†–´ –ò–ó –£–°–õ–û–í–ò–Ø –ó–ê–î–ê–ß–ò.

# –ò–ó–ú–ï–ù–ï–ù–ò–Ø:
eta = 1e-2
n_iter = 2000
# –ò–ó–ú–ï–ù–ï–ù–ò–Ø

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if i % 10 == 0:
#       –ò–ó–ú–ï–ù–ï–ù–ò–Ø
        eta /= 1.005
#       –ò–ó–ú–ï–ù–ï–ù–ò–Ø 
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
```

```python
# –í—ã–≤–æ–¥: –Ω–∞–∏–ª—É—á—à–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–æ–ª—É—á–∞–µ—Ç—Å—è –ø—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∏—Ç–µ—Ä–∞—Ü–∏–π (–¥–æ 2 000) –∏ 
# –ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ –¥—Ä—É–≥–æ–≥–æ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∏–∑–º–µ–Ω–µ–Ω–∏—è —Å–∫–æ—Ä–æ—Å—Ç–∏ –æ–±—É—á–µ–Ω–∏—è (1.005)- —Å—Ç—Ä–æ–∫–∞ 23.
# –î–∞–ª—å–Ω–µ–π—à–µ–≥–æ —É–ª—É—á—à–µ–Ω–∏—è MSE –Ω–µ –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è.
```

**2*. –í —ç—Ç–æ–º –∫–æ–¥–µ –º—ã –∏–∑–±–∞–≤–ª—è–µ–º—Å—è –æ—Ç –∏—Ç–µ—Ä–∞—Ü–∏–π –ø–æ –≤–µ—Å–∞–º, –Ω–æ —Ç—É—Ç –µ—Å—Ç—å –æ—à–∏–±–∫–∞, –∏—Å–ø—Ä–∞–≤—å—Ç–µ –µ–µ.**

```python
# –û—à–∏–±–∫–∞ –≤–æ–∑–Ω–∏–∫–∞–µ—Ç –ø–æ –ø—Ä–∏—á–∏–Ω–µ —Ç–æ–≥–æ, —á—Ç–æ –º–∞—Ç—Ä–∏—Ü–∞ –• –∏–º–µ–µ—Ç —Ä–∞–∑–º–µ—Ä 10—Ö2, –∞ y_pred-y –∏–º–µ–µ—Ç —Ä–∞–∑–º–µ—Ä 10—Ö1.
# –î–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–ª—è —É–º–Ω–æ–∂–µ–Ω–∏—è –≤–∑—è—Ç—å –∏–∑ –• —Ç–æ–ª—å–∫–æ —Å—Ç–æ–ª–±–µ—Ü –∑–Ω–∞—á–µ–Ω–∏–π - [:, 1], 
# –Ω–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–π –±—É–¥–µ—Ç —Ñ–æ—Ä–º—É–ª–∞ –≤–∏–¥–∞: W -= eta * (2/n * X.T @ (X @ W - y)).
# –í —ç—Ç–æ–º —Å–ª—É—á–∞–µ –º—ã –ø–æ–ª—É—á–∏–º MSE –∫–∞–∫ –∏ –≤ –ø—Ä–µ–¥—ã–¥—É—â–µ–º –ø—Ä–∏–º–µ—Ä–µ.
```

```python
n = X.shape[0]

eta = 1e-2 
n_iter = 1000

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
#     for k in range(W.shape[0]):
#         W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    # –ò–ó–ú–ï–ù–ï–ù–ò–Ø
    W -= eta * (2/n * X.T @ (X @ W - y))
    # –ò–ó–ú–ï–ù–ï–ù–ò–Ø
    #
    if i % 10 == 0:
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err,2)}')
```
