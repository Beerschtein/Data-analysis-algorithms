**1. Подберите скорость обучения (eta) и количество итераций**

```python
# Задача: предсказание баллов ЕГЭ ученика в зависимости от кол-ва лет стажа его репетитора
```

```python
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
```

```python
X.shape
```

```python
y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
```

```python
# Напишем функцию для подсчета среднеквадратичной ошибки:
```

```python
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

```python
# Имеется модель, определяющая "веса" w для неизвестной x методом градиентного спуска:
```

```python
n = X.shape[0]

# eta = 1e-2
# n_iter = 100 - ПАРАМЕТРЫ ИЗ УСЛОВИЯ ЗАДАЧИ.

# ИЗМЕНЕНИЯ:
eta = 1e-2
n_iter = 2000
# ИЗМЕНЕНИЯ

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if i % 10 == 0:
#       ИЗМЕНЕНИЯ
        eta /= 1.005
#       ИЗМЕНЕНИЯ 
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
```

```python
# Вывод: наилучшй результат получается при увеличении количества итераций (до 2 000) и 
# при использовании другого коэффициента изменения скорости обучения (1.005)- строка 23.
# Дальнейшего улучшения MSE не наблюдается.
```

**2*. В этом коде мы избавляемся от итераций по весам, но тут есть ошибка, исправьте ее.**

```python
# Ошибка возникает по причине того, что матрица Х имеет размер 10х2, а y_pred-y имеет размер 10х1.
# Достаточно для умножения взять из Х только столбец значений - [:, 1], 
# но правильной будет формула вида: W -= eta * (2/n * X.T @ (X @ W - y)).
# В этом случае мы получим MSE как и в предыдущем примере.
```

```python
n = X.shape[0]

eta = 1e-2 
n_iter = 1000

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
#     for k in range(W.shape[0]):
#         W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    # ИЗМЕНЕНИЯ
    W -= eta * (2/n * X.T @ (X @ W - y))
    # ИЗМЕНЕНИЯ
    #
    if i % 10 == 0:
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err,2)}')
```
